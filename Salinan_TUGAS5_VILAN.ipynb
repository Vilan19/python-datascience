```python
from google.colab import files
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DATASET/earthquakes.csv')
df
```

```python
df_copy = df.copy()
df_copy
```

```python
# Ambil kolom yang diinginkan
df2 = df_copy[['id', 'magnitude', 'type', 'title', 'date', 'time', 'alert', 'status', 'depth', 'latitude', 'longitude', 'location', 'country']]
df2
```

```python
# lambda
# Tambahkan kolom klasifikasi menggunakan lambda function
df2.loc[:,'klasifikasi'] = df2['magnitude'].apply( # menerapkan fungsi ke setiap elemen
    lambda x: 'Micro' if x < 2.0 else
              'Minor' if 2.0 <= x < 4.0 else
              'Light' if 4.0 <= x < 5.0 else
              'Moderate' if 5.0 <= x < 6.0 else
              'Strong' if 6.0 <= x < 7.0 else
              'Major' if 7.0 <= x < 8.0 else
              'Great'
)
df3 = df2[['magnitude', 'klasifikasi']]
df3
```

```python
# Rekursi
import sys
sys.setrecursionlimit(1400)  # Ubah batas rekursi

def max_mag(df3, index, current_max):
    # Basis kasus: jika index di luar batas
    if index < 0:
        return current_max

    # Ambil magnitudo dari baris
    current_mag = df3.iloc[index]['magnitude'] if pd.notnull(df3.iloc[index]['magnitude']) else float('-inf')

    # Perbarui current_max jika current_magnitude lebih tinggi
    current_max = max(current_max, current_mag)

    # Panggil rekursi untuk index sebelumnya
    return max_mag(df3, index - 1, current_max)

# Mencari magnitudo tertinggi dari semua gempa
highest_magnitude = max_mag(df3, len(df3) - 1, float('-inf'))

print(f'Magnitudo tertinggi: {highest_magnitude}')
```

```python
# First class
# Fungsi untuk menghitung jumlah gempa berdasarkan kriteria yang ditentukan oleh fungsi lain
def jumlah_gempa(df3, kriteria):
    count = 0
    for i in range(len(df3)):
        if kriteria(df3.iloc[i]):  # Memanggil fungsi kriteria untuk setiap gempa
            count += 1
    return count

# Fungsi kriteria untuk magnitudo lebih besar dari 5.0
def high_magnitude(gempa):
    return gempa['magnitude'] > 5.0  # Menggunakan kolom 'magnitude'

# Menggunakan fungsi first-class
total_gempa = jumlah_gempa(df3, high_magnitude)

print(f'Jumlah gempa dengan magnitudo lebih dari 5.0 adalah {total_gempa}')
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of countries to analyze
countries = ['India', 'USA', 'Indonesia', 'China', 'Afghanistan']

# Filter the data for the selected countries
df_filtered = df2[df2['country'].isin(countries)]

# Count the number of earthquakes per country
earthquake_counts = df_filtered['country'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=earthquake_counts.index, y=earthquake_counts.values)
plt.title('Perbandingan Frekuensi Gempa Negara USA, Indonesia, China, Afghanistan dan India', fontsize=10)
plt.xlabel('Nama Negara', fontsize=10)
plt.ylabel('Frekuensi Gempa', fontsize=10)
plt.show()
```

```python
# Mengelompokkan data gempa berdasarkan negara dan menghitung jumlah kejadian
jumlah_gempa = df2['country'].value_counts().head(5)

# Mengambil data hanya untuk 10 negara dengan gempa terbanyak
df_top_countries = df2[df2['country'].isin(jumlah_gempa.index)]

plt.figure(figsize=(10, 6))
# Looping melalui setiap negara untuk scatter plot
for country in jumlah_gempa.index:
    data_negara = df_top_countries[df_top_countries['country'] == country]
    plt.scatter(data_negara['country'], data_negara['magnitude'], label=country, s=50)

# Menyesuaikan label dan judul
plt.title('Top 5 Countries with the Most Earthquakes and Their Magnitudes', fontsize=14)
plt.xlabel('Country', fontsize=10)
plt.ylabel('Magnitude of Earthquakes', fontsize=10)

# Menampilkan legenda
plt.legend(title='Country')

# Menampilkan plot
plt.show()

```
